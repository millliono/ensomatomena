```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1p_1c = pd.read_csv('1p_1c.csv', header=None)
df_1p_2c = pd.read_csv('1p_2c.csv', header=None)
df_1p_3c = pd.read_csv('1p_3c.csv', header=None)
df_2p_1c = pd.read_csv('2p_1c.csv', header=None)
df_2p_2c = pd.read_csv('2p_2c.csv', header=None)
df_2p_3c = pd.read_csv('2p_3c.csv', header=None)
df_3p_1c = pd.read_csv('3p_1c.csv', header=None)
df_3p_2c = pd.read_csv('3p_2c.csv', header=None)
df_3p_3c = pd.read_csv('3p_3c.csv', header=None)
```

```python
y=[]
y.append(df_1p_1c.mean(axis=1).values[0])
y.append(df_1p_2c.mean(axis=1).values[0])
y.append(df_1p_3c.mean(axis=1).values[0])
y.append(df_2p_1c.mean(axis=1).values[0])
y.append(df_2p_2c.mean(axis=1).values[0])
y.append(df_2p_3c.mean(axis=1).values[0])
y.append(df_3p_1c.mean(axis=1).values[0])
y.append(df_3p_2c.mean(axis=1).values[0])
y.append(df_3p_3c.mean(axis=1).values[0])

```

```python
x = ['1p_1c', '1p_2c', '1p_3c', '2p_1c', '2p_2c', '2p_3c', '3p_1c', '3p_2c', '3p_3c']
plt.ylabel("time (us)")
for i, v in enumerate(y):
    plt.text(i, v+1, str(int(v)), ha='center')
plt.bar(x, y)
```
